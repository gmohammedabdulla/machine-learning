```python
from modular_ann import *
import numpy
import matplotlib.pyplot as plt
%matplotlib inline  
%load_ext autoreload
%autoreload 2
```

```python
#Creating a data set
l = 500
X1 = np.random.randn(l, 2) + np.array([0,2])
X2 = np.random.randn(l, 2) + np.array([0,-2])
X3 = np.random.randn(l, 2) + np.array([-2,0])

Y = np.array([0]*l + [1]*l + [2]*l)
X = np.vstack([X1, X2, X3])

assert(len(X) == len(Y))

```

```python
#Visualise the dataset
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
T = one_hot_encode(Y)
```

```python
N = 5
K = 3
D = 2
W1 = np.random.randn(D, N)
b1 = np.random.randn(N)
W2 = np.random.randn(N, K)
b2 = np.random.randn(K)
#Full Gradient decent
alpha = 0.0001
full_cost_list = []
epochs = 100
for i in range(epochs):
    Y_hat, Z = feed_forward(X, W1, b1 , W2, b2)
    dW2, db2, dZ = derivative_l2(Y_hat, T, Z, W2)
    dW1, db1 = derivative_l1(dZ, Z, X)
    cost = cost_function(Y_hat, T)
    full_cost_list.append(cost)
    
    W1 = W1 - alpha * dW1 
    W2 = W2 - alpha * dW2
    b1 = b1 - alpha * db1
    b2 = b2 - alpha * db2

P_Y_given_X, Z = feed_forward(X, W1, b1, W2, b2)
print P_Y_given_X.shape
P = np.argmax(P_Y_given_X, axis=1)
print np.sum(Y == P) * 1.0/ len(Y)
```

```python
N = 5
K = 3
D = 2
W1 = np.random.randn(D, N)
b1 = np.random.randn(N)
W2 = np.random.randn(N, K)
b2 = np.random.randn(K)
#stocastic Gradient decent
alpha = 0.0001
stocastic_cost_list = []
epochs = 100
batch_size = 128
for i in range(epochs):
    for j in range(1500/batch_size):
        s = j * batch_size
        X_small = X[s:s+batch_size]
        T_small = T[s:s+batch_size]
        Y_hat, Z = feed_forward(X_small, W1, b1 , W2, b2)
        dW2, db2, dZ = derivative_l2(Y_hat, T_small, Z, W2)
        dW1, db1 = derivative_l1(dZ, Z, X_small)
        
        W1 = W1 - alpha * dW1 
        W2 = W2 - alpha * dW2
        b1 = b1 - alpha * db1
        b2 = b2 - alpha * db2
    
    cost = cost_function(Y_hat, T_small)
    stocastic_cost_list.append(cost)
    
    

P_Y_given_X, Z = feed_forward(X, W1, b1, W2, b2)
print P_Y_given_X.shape
P = np.argmax(P_Y_given_X, axis=1)
print np.sum(Y == P) * 1.0/ len(Y)
```
